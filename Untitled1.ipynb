```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt
%matplotlib inline


x = np.arange(0.0, 1.0, 0.01)
print(len(x))
y = np.sin(np.pi * x + 0.5)
plt.plot(x, y)
plt.show()
  
n = len(y)
```

```python
freqs = np.fft.rfft(y)
magn = np.sqrt(freqs.imag**2 + freqs.real**2)
plt.plot(range(len(freqs))[:100], magn)
plt.show()
print(magn[:10])
```

```python
y_new = np.fft.irfft(freqs)
plt.plot(x, y_new)
plt.plot(x, y, color="red")
plt.show()
```
